```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("C:\\Users\\jinhj\\OneDrive\\바탕 화면\\연대의대\\예2\\초연멘\\14주차\\data_set_isFiles_result.csv")
dataset
```

```python
plt.title('Number of patients by age')
plt.xlabel('Age')
plt.ylabel('Number of patients')
plt.hist(dataset['age'].tolist(), bins=8)
plt.show()
```

```python
sex_list=dataset['sex'].tolist()
total_num_male=sex_list.count(0)
total_num_female=sex_list.count(1)
ratio=[total_num_male, total_num_female]
labels=['Male', 'Female']
colors = ['#307EC9', '#FB627E']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
plt.pie(ratio, labels=labels,autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
total_num_male
```

```python
plt.figure(figsize=(12, 6))
num_of_patients_by_diagnosis_male=dataset[dataset['sex']==0].groupby('diagnosis').diagnosis.count()
num_of_patients_by_diagnosis_female=dataset[dataset['sex']==1].groupby('diagnosis').diagnosis.count()

label=['ADHD-Combined Type','ADHD-Hyperactive/Impulsive Type', 'ADHD-Inattentive Type', 'No Diagnosis Given']
n=len(label)
index=np.arange(n)
bar_width=0.35
alpha=0.9
p1=plt.bar(index, num_of_patients_by_diagnosis_male, bar_width, color='#307EC9', alpha=alpha, label='Male')
p2=plt.bar(index+bar_width, num_of_patients_by_diagnosis_female, bar_width, color='#FB627E', alpha=alpha, label='Female')
plt.title('Number of patients by diagnosis & sex')
plt.xlabel('Diagnosis',fontsize=9)
plt.ylabel('Number of patients')
plt.xticks(index,label,fontsize=7)
plt.legend((p1[0], p2[0]),('Male','Female'), fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
num_of_patients_by_diagnosis_5to9=dataset[dataset['age_group']=='5~9'].groupby('diagnosis').diagnosis.count()
num_of_patients_by_diagnosis_10to14=dataset[dataset['age_group']=='10~14'].groupby('diagnosis').diagnosis.count()
num_of_patients_by_diagnosis_15or_older=dataset[dataset['age_group']=='15~'].groupby('diagnosis').diagnosis.count()

'add count=0'
num_of_patients_by_diagnosis_15or_older.loc['ADHD-Hyperactive/Impulsive Type']=0
num_of_patients_by_diagnosis_15or_older=num_of_patients_by_diagnosis_15or_older.reindex(index=['ADHD-Combined Type','ADHD-Hyperactive/Impulsive Type', 'ADHD-Inattentive Type', 'No Diagnosis Given'])

label=['ADHD-Combined Type','ADHD-Hyperactive/Impulsive Type', 'ADHD-Inattentive Type', 'No Diagnosis Given']
n=len(label)
index=np.arange(n)
bar_width=0.25
alpha=0.9
p1=plt.bar(index-bar_width, num_of_patients_by_diagnosis_5to9, bar_width, color='#307EC9', alpha=alpha, label='5~9')
p2=plt.bar(index, num_of_patients_by_diagnosis_10to14, bar_width, color='#FB627E', alpha=alpha, label='10~14')
p3=plt.bar(index+bar_width, num_of_patients_by_diagnosis_15or_older, bar_width, color='#F2F0A1', alpha=alpha, label='15~21')
plt.title('Number of patients by diagnosis & age')
plt.xlabel('Diagnosis',fontsize=9)
plt.ylabel('Number of patients')
plt.xticks(index,label,fontsize=7)
plt.legend((p1[0], p2[0], p3[0]),('5~9','10~14','15~'), fontsize=15)
plt.show()
num_of_patients_by_diagnosis_15or_older
```

```python
plt.figure(figsize=(12, 6))
num_of_patients_by_age_ADHD_CombinedType=dataset[dataset['diagnosis']=='ADHD-Combined Type'].groupby('age_group').age_group.count()
num_of_patients_by_age_ADHD_Hyperactive_ImpulsiveType=dataset[dataset['diagnosis']=='ADHD-Hyperactive/Impulsive Type'].groupby('age_group').age_group.count()
num_of_patients_by_age_ADHD_InattentiveType=dataset[dataset['diagnosis']=='ADHD-Inattentive Type'].groupby('age_group').age_group.count()
num_of_patients_by_age_ADHD_NoDiagnosisGiven=dataset[dataset['diagnosis']=='No Diagnosis Given'].groupby('age_group').age_group.count()

'add count=0, reindex'
num_of_patients_by_diagnosis_15or_older.loc['15~']=0
num_of_patients_by_age_ADHD_CombinedType=num_of_patients_by_age_ADHD_CombinedType.reindex(index=['5~9', '10~14', '15~'])
num_of_patients_by_age_ADHD_Hyperactive_ImpulsiveType=num_of_patients_by_age_ADHD_Hyperactive_ImpulsiveType.reindex(index=['5~9', '10~14', '15~'])
num_of_patients_by_age_ADHD_InattentiveType=num_of_patients_by_age_ADHD_InattentiveType.reindex(index=['5~9', '10~14', '15~'])
num_of_patients_by_age_ADHD_NoDiagnosisGiven=num_of_patients_by_age_ADHD_NoDiagnosisGiven.reindex(index=['5~9', '10~14', '15~'])




labels=['5~9','10~14', '15~',]
n=len(labels)
index=np.arange(n)
bar_width=0.2
alpha=0.9
p1=plt.bar(index-bar_width*1.5, num_of_patients_by_age_ADHD_CombinedType, bar_width, color='#307EC9', alpha=alpha, label=labels[0])
p2=plt.bar(index-bar_width*0.5, num_of_patients_by_age_ADHD_Hyperactive_ImpulsiveType, bar_width, color='#FB627E', alpha=alpha, label=labels[1])
p3=plt.bar(index+bar_width*0.5, num_of_patients_by_age_ADHD_InattentiveType, bar_width, color='#F2F0A1', alpha=alpha, label=labels[2])
p4=plt.bar(index+bar_width*1.5, num_of_patients_by_age_ADHD_NoDiagnosisGiven, bar_width, color='#BF9BDE', alpha=alpha, label=label[3])
plt.title('Number of patients by age & diagnosis')
plt.xlabel('age group',fontsize=15)
plt.ylabel('Number of patients')
plt.xticks(index,labels,fontsize=15)
plt.legend((p1[0], p2[0], p3[0], p4[0]),('ADHD-Combined Type','ADHD-Hyperactive/Impulsive Type', 'ADHD-Inattentive Type', 'No Diagnosis Given'), fontsize=10, loc=1)
plt.show()
```
